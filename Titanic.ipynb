```python
# preprocessing data(cleaning and organising)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df["Sex"]=le.fit_transform(df["Sex"])
df
```

```python
df["Embarked"]=le.fit_transform(df["Embarked"])
```

```python
# input data
x=df.drop(['Survived'],axis=1).values
x
```

```python
y=df['Survived'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.fit_transform(x_train)
x_test=scalar.fit_transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbours_model=KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.svm import SVC
svm_model=SVC()
```

```python
# perfomance evaluation techniqes
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay,classification_report
model=[neighbours_model,svm_model]
for i in model:
  i.fit(x_train,y_train)
  y_prede=i.predict(x_test)
  result=confusion_matrix(y_test,y_prede)
  score=accuracy_score(y_test,y_prede)
  report=classification_report(y_test,y_prede)
  cm=['alive', 'dead']
  cmd=ConfusionMatrixDisplay(result,display_labels=cm)
  print("model name :- ",i,"\n",y_prede)
  print(result)
  print(score)
  print(report)
  print(cmd.plot())
  print("*"*100)
```
